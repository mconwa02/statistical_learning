```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
College = sm.datasets.get_rdataset('College', 'ISLR')
```

```python
College.title
```

```python
df = pd.DataFrame(College.data)
```

```python
df.index
```

```python
df.info()
```

```python
df.describe()
```

```python
pd.plotting.scatter_matrix(df, figsize=(15, 15))
plt.tight_layout()
plt.show()
```

```python
df[['Outstate', 'Private']].boxplot(by='Private')
plt.show()
```

```python
df['Elite'] = ['Yes' if x >= 50 else 'No' for x in df.Top10perc]
```

```python
df.Elite.value_counts()
```

```python
df[['Outstate', 'Elite']].boxplot(by='Elite')
plt.show()
```

```python
pd.plotting.scatter_matrix(df[['Top10perc', 'Top25perc', 'Room.Board', 'Personal']], figsize=(15, 15))
plt.tight_layout()
plt.show()
```
